```python
import requests
```

# setting scrapping website 

```python
url = "https://www.flipkart.com/search?q=watches&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_5_0_na_na_na&as-pos=5&as-type=TRENDING&suggestionId=watches&requestId=455814f8-290a-42b2-a063-e6661b29d03b"
```

```python
allow = requests.get(url)
```

```python
print(allow)
```

# testing scrapping allowability

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

```python
cast = soup.select(".IRpwTa")
```

```python
cast
```

```python
tinky=[]
```

```python
for item in cast:
    tinky.append(item.text)
```
